```python
import pandas as pd
import datetime as dt
Real_Prop = pd.read_csv("../data/EXTR_RPSale.csv")
Res_Bldg = pd.read_csv("../data/EXTR_ResBldg.csv")
Parcel = pd.read_csv("../data/EXTR_Parcel.csv", encoding='latin-1')
codes = pd.read_csv('../data/EXTR_LookUp.csv')
```

Since we are only interested in 2019 data, using datetime to convert documentdate and filtering for 2019 sales

```python
Real_Prop['DocumentDate'] = pd.to_datetime(Real_Prop['DocumentDate'])
Real_Prop19 = Real_Prop[Real_Prop['DocumentDate'].dt.year == 2019]
Real_Prop19
```

formatting column names to make them more Pythonic!

```python
Real_Prop19 = Real_Prop19.rename(columns=str.lower)
```

```python
Real_Prop19
```

padding major and minor and adding them together after converting them to string type

```python
Real_Prop19['major'] = Real_Prop19['major'].astype(str)
Real_Prop19['minor'] = Real_Prop19['minor'].astype(str)
```

```python
def major_padded(row):
    if len(row) == 6:
        return row
    elif len(row) == 5:
        return '0'+row
    elif len(row) == 4:
        return '00'+row
    elif len(row) == 3:
        return '000'+row
    elif len(row) == 2:
        return '0000'+row
    elif len(row) == 1:
        return '00000'+row
    else:
        return '000000'

def minor_padded(row):
    if len(row) == 4:
        return row
    elif len(row) == 3: 
        return '0'+row
    elif len(row) == 2:
        return '00'+row
    elif len(row) == 1:
        return '000'+row
    else:
        return '0000'
    
Real_Prop19['major'] = Real_Prop19.major.apply(major_padded)
Real_Prop19['minor'] = Real_Prop19.minor.apply(minor_padded)
Real_Prop19['major_minor'] = Real_Prop19['major']+Real_Prop19['minor']
```

```python
Real_Prop19
```

moving the major_minor column to column1 position

```python
col_name = 'major_minor'
last_col = Real_Prop19.pop(col_name)
Real_Prop19.insert(1, col_name, last_col)
```
